```python
df_clean=df.dropna()

print('Rows:',df_clean.shape[0])
print('Columns:',df_clean.shape[1])
```

```python
df_clean=df_clean.filter(items=["title","genres","rating","votes","release_date"])
print(df_clean)
```

```python
def clean_votes(v):
    if pd.isna(v):
        return 0
    v = v.replace(',', '')
    if 'K' in v:
        return int(float(v.replace('K','')) * 1000)
    elif 'M' in v:
        return int(float(v.replace('M','')) * 1_000_000)
    else:
        return int(v)

# Correct way: assign back to the column
df_clean['votes'] = df_clean['votes'].apply(clean_votes)

print(df_clean.head())
```

```python
df_clean['year']=pd.to_datetime(df_clean['release_date']).dt.year
print(df_clean.head())
```

```python


df_clean['genres'] = df_clean['genres'].fillna('').str.replace(r"[\[\]'\"']", '', regex=True).str.split(',')
df_clean = df_clean.explode('genres').reset_index(drop=True)


df_clean['genres'] = df_clean['genres'].str.strip()
df_clean = df_clean[['title','genres','year','votes','rating']]

print(df_clean.head(10))
```

```python
#numerical analysis
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
print(df[numeric_cols].describe())
```

```python
#highest and least ratings
print(f"Mean Rating: {df['rating'].mean():.2f}")
print(f"Median Rating: {df['rating'].median():.2f}")
print("Highest rated movie:")
print(df[df['rating']==df['rating'].max()][['title','rating']])
print("Lowest rated movie:")
print(df[df['rating']==df['rating'].min()][['title','rating']])
```

```python
print(df['votes'].dtype)
print(df['votes'].head(10))
```

```python
#to convert votes into int dtype
def votes_to_int(vote_str):
    if pd.isna(vote_str):
        return 0
    vote_str = str(vote_str).strip().lower()
    if 'k' in vote_str:
        return int(float(vote_str.replace('k','')) * 1000)
    elif 'm' in vote_str:
        return int(float(vote_str.replace('m','')) * 1_000_000)
    else:
        try:
            return int(vote_str)
        except ValueError:
            return 0
df['votes'] = df['votes'].apply(votes_to_int)
print(df['votes'].head())
print(df['votes'].isna().sum())
```

```python
#release_date/year
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_date'] = df['release_date'].fillna(pd.Timestamp('1900-01-01'))
df['release_year'] = df['release_date'].dt.year.astype(int)
print(df['release_year'].head())
print("Earliest movie year:", df['release_year'].min())
print("Most recent movie year:", df['release_year'].max())
print("Year with most movies:")
print(df['release_year'].value_counts().idxmax(), "-", df['release_year'].value_counts().max(), "movies")
```

```python
#Categorical Analysis
print("Categorical Analysis")
print("\nGenres Count ")
print(df['genres'].value_counts())
print("\n Ratings by Genres")
genres_rating = df.groupby('genres')['rating'].agg(['count','mean','min','max']).sort_values('mean', ascending=False)
print(genres_rating)

# Year-wise rating trends
print("\nRelease Year-wise Rating Trends")
release_rating = df.groupby('release_year')['rating'].agg(['count','mean']).sort_index()
print(release_rating)
```

```python
#movies with rating >=1000
print("\nTop Movies")
top_movies = df[df['votes'] >= 1000].sort_values(by='rating', ascending=False).head(10)
print(top_movies[['title','genres','release_year','rating','votes']])
```

```python
# duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
# Correlation matrix
correlation_matrix = df.corr(numeric_only=True)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
#IQR
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
Q1 = df['budget'].quantile(0.25)
Q3 = df['budget'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['budget'] < lower_bound) | (df['budget'] > upper_bound)]
print("Number of budget outliers:", outliers.shape[0])
```
